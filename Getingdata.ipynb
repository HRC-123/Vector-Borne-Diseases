```python
import json
import pandas as pd
import requests
import csv
```

```python

location = 'jalandhar'
key1 = 'EB36A28R7CW4FW3M5PW4KP5Z8'
key2 = '8HNQJQGMZQKPRPGN3T6UAM9KV'

for i in range(2017, 2021):
    from_date = f'{i}-01-01'
    to_date = f'{i}-12-31'

    response = 'NULL'
    if(i%2==0):
        response = requests.get(f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{location}/{from_date}/{to_date}?unitGroup=uk&include=days&key={key1}&contentType=json')
    else:
        response = requests.get(f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{location}/{from_date}/{to_date}?unitGroup=uk&include=days&key={key2}&contentType=json')
    
    

    if response.status_code == 200:
        # Extracting data from response
        data = response.json()["days"]

        # Save data to JSON file
        with open(f"{location}_response_{i}.json", "w") as f:
            json.dump(data, f)
```

```python

# Initialize an empty list to store data from JSON files
all_data = []

# Read data from JSON files
for i in range(2017, 2021):
    filename = f"{location}_response_{i}.json"
    with open(filename, "r") as f:
        data = json.load(f)
        all_data.extend(data)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(all_data)

# Convert 'datetime' column to datetime type
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract year and month from datetime
df['Year'] = df['datetime'].dt.year
df['Month'] = df['datetime'].dt.month

# Calculate average values for dew, humidity, and precip for each month and year
avg_values_df = df.groupby(['Year', 'Month'])[['temp', 'dew', 'humidity', 'precip']].mean().reset_index()

# Rename columns
avg_values_df.columns = ['Year', 'Month', 'Avg_Temp', 'Avg_Dew', 'Avg_Humidity', 'Avg_Precipitation']

# Save to CSV
avg_values_df.to_csv(f"{location}_average_weather.csv", index=False)

print(avg_values_df)
```

```python

# location = "amritsar"
def csv_to_json_with_district(csv_file_path, json_file_path, district_name):
    # Read the CSV file
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        # Convert CSV to a list of dictionaries
        data = list(csv_reader)

    # Create a dictionary with district name and data
    json_data = {"District": district_name, "data": data}

    # Write data to a JSON file
    with open(json_file_path, 'w') as json_file:
        json.dump(json_data, json_file, indent=4)

# Example usage:
csv_file_path = f"{location}_average_weather.csv"
json_file_path = f"{location}_average_weather.json"
district_name = f"{location}"
csv_to_json_with_district(csv_file_path, json_file_path, district_name)
```
