```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Read Cases data
cases_data = pd.read_csv('updated_file.csv')

# Filter data for Ludhiana district
ludhiana_cases = cases_data[cases_data['District'] == 'Ludhiana']

# Read Weather data
weather_data = pd.read_csv('ludhiana_average_weather.csv')

# Merge Cases and Weather data on 'Month' and 'Year' columns
merged_data = pd.merge(ludhiana_cases, weather_data, left_on=['Month'], right_on=['Month'])

print(merged_data.head(20))

# Define features and target variable
X = merged_data[['Avg_Temp', 'Avg_Dew', 'Avg_Humidity', 'Avg_Precipitation']]
y = merged_data[['2017', '2018', '2019', '2020']]

# Initialize and fit the regression model
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print("Coefficients:")
for year, coef in zip(y.columns, model.coef_):
    print(year + ": ", coef)

# Print intercept
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import r2_score

# Predict the target variable
y_pred = model.predict(X)

# Calculate R-squared
r_squared = r2_score(y, y_pred)
print("R-squared:", r_squared)
```

```python
from sklearn.metrics import mean_squared_error

# Calculate Mean Squared Error
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Function to perform regression for a single district
def perform_regression(district, cases_file, weather_file):
    # Read cases data for the specific district
    cases_data = pd.read_csv(cases_file)
    district_cases = cases_data[cases_data['District'] == district]
    
    # Read weather data for the specific district
    weather_data = pd.read_csv(weather_file)
    
    # Merge cases and weather data
    merged_data = pd.merge(district_cases, weather_data, on='Month')
    
    print(merged_data.head())
    # Define features and target variable
    X = merged_data[['Avg_Temp', 'Avg_Dew', 'Avg_Humidity', 'Avg_Precipitation']]
    y = merged_data[['2017', '2018', '2019', '2020']]
    
    # Initialize and fit the regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Predict the target variable
    y_pred = model.predict(X)
    
    # Calculate R-squared
    r_squared = r2_score(y, y_pred)
    
    # Calculate Mean Squared Error
    mse = mean_squared_error(y, y_pred)
    
    return model, r_squared, mse

# Districts and corresponding files
districts = ['Amritsar', 'Barnala', 'Ludhiana', 'Mansa', 'Patiala']
cases_files = ['updated_file.csv'] * len(districts)
weather_files = ['amritsar_average_weather.csv', 'barnala_average_weather.csv', 
                 'ludhiana_average_weather.csv', 'mansa_average_weather.csv', 
                 'patiala_average_weather.csv']

# Perform regression for each district
for district, cases_file, weather_file in zip(districts, cases_files, weather_files):
    print("District:", district)
    model, r_squared, mse = perform_regression(district, cases_file, weather_file)
    print("R-squared:", r_squared)
    print("Mean Squared Error:", mse)
    print()
```

```python
import pandas as pd

# Load district-wise cases data
cases_data = pd.read_csv('modified_file.csv')

# Initialize an empty list to store individual district dataframes
dfs = []

# List of districts
districts = ['amritsar', 'barnala', 'bathinda', 'faridkot', 'fatehgarhsahib', 'fazilka', 'ferozepur', 'gurdaspur', 'hoshiarpur', 'jalandhar', 'kapurthala', 'ludhiana', 'mansa', 'moga', 'pathankot', 'patiala', 'rupnagar', 'sahibzadaajitsinghnagar(mohali)', 'nawanshahr', 'sangrur', 'shrmukatsarsahib', 'tarntaran']

# Loop through each district
for district in districts:
    # Load weather data for the current district
    weather_data = pd.read_csv(f'./Weather/{district}/{district}_average_weather.csv')
    
    # Merge weather data with cases data based on District and Month
    merged_data = pd.merge(cases_data[cases_data['District'] == district], weather_data, on='Month')
    
    # Reshape merged data to long format
    merged_data = pd.melt(merged_data, id_vars=['District', 'Month'], var_name='Year', value_name='Cases')
    
    # Append merged dataframe to the list
    dfs.append(merged_data)

# Concatenate all district dataframes into a single dataframe
final_data = pd.concat(dfs)

# Remove any non-numeric values from the 'Year' column
final_data = final_data[final_data['Year'].str.isnumeric()]

# Convert 'Year' column to int64
final_data['Year'] = final_data['Year'].astype(int)

# Merge with original weather data to get other weather attributes
final_data = pd.merge(final_data, weather_data, on=['Year', 'Month'])

# Select relevant columns
final_data = final_data[['District', 'Year', 'Month', 'Cases', 'Avg_Temp', 'Avg_Dew', 'Avg_Humidity', 'Avg_Precipitation']]

# Remove duplicate rows
final_data = final_data.drop_duplicates()

# Save the final merged data to a CSV file
final_data.to_csv("merged_data.csv", index=False) 

# Display the final dataframe
print(final_data)
```
