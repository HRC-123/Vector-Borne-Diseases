```python

# Initialize an empty list to store data from JSON files
all_data = []

# Read data from JSON files
for i in range(2017, 2021):
    filename = f"{location}_response_{i}.json"
    with open(f"./Weather/{location}/{filename}", "r") as f:
        data = json.load(f)
        all_data.extend(data)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(all_data)

# Convert 'datetime' column to datetime type
df['datetime'] = pd.to_datetime(df['datetime'])

# Extract year and month from datetime
df['Year'] = df['datetime'].dt.year
df['Month'] = df['datetime'].dt.month

# Calculate average values for dew, humidity, and precip for each month and year
avg_values_df = df.groupby(['Year', 'Month'])[['temp','feelslike','dew','humidity','precip','precipprob','precipcover','snow','snowdepth','windgust','windspeed','winddir','pressure','cloudcover',	'visibility','solarradiation','solarenergy','uvindex']].mean().reset_index()

# Rename columns
avg_values_df.columns = ['Year', 'Month','Avg_Temp', 'Avg_Feelslike', 'Avg_Dew', 'Avg_Humidity', 'Avg_Precipitation', 'Avg_Precipitation_Probability', 'Avg_Precipitation_Coverage', 'Avg_Snowfall', 'Avg_Snow_Depth', 'Avg_Wind_Gust', 'Avg_Wind_Speed', 'Avg_Wind_Direction', 'Avg_Pressure', 'Avg_Cloud_Cover', 'Avg_Visibility', 'Avg_Solar_Radiation', 'Avg_Solar_Energy', 'Avg_UV_Index']

# Save to CSV
avg_values_df.to_csv(f"./Weather/{location}/{location}_average_weather.csv", index=False)

print(avg_values_df)
```

```python

# location = "amritsar"
def csv_to_json_with_district(csv_file_path, json_file_path, district_name):
    # Read the CSV file
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        # Convert CSV to a list of dictionaries
        data = list(csv_reader)

    # Create a dictionary with district name and data
    json_data = {"District": district_name, "data": data}

    # Write data to a JSON file
    with open(json_file_path, 'w') as json_file:
        json.dump(json_data, json_file, indent=4)

# Example usage:
csv_file_path = f"./Weather/{location}/{location}_average_weather.csv"
json_file_path = f"./Weather/{location}/{location}_average_weather.json"
district_name = f"{location}"
csv_to_json_with_district(csv_file_path, json_file_path, district_name)
```
